**Este código demuestra lo sencillo que es construir un modelo de regresión lineal simple utilizando conceptos estadísticos. A partir de un conjunto de datos conocidos de las variables X e Y, se calcula una ecuación lineal que describe la relación entre ambas variables. Una vez obtenido el modelo, podemos introducir nuevos valores de X y predecir los valores correspondientes de Y, incluso cuando estos son desconocidos.**

```python
#Importamos las librerias
import numpy as np
```

```python
# Variable independiente (X)
X = np.array([1,2,3,4,5])
```

```python
# Variable dependiente (Y)
Y = np.array([ 2,4,5,4,5])
```

```python
# Cantidad de observaciones
N = len(X)
```

```python
# Sumatorias necesarias para la fórmula
X_sum = np.sum(X)
Y_sum = np.sum(Y)
XY_sum = np.sum(X * Y)
X2_sum = np.sum(X ** 2)
```

```python
# Cálculo de la pendiente (B)
B = (N * XY_sum - X_sum * Y_sum) / (N * X2_sum - X_sum**2)
```

```python
# Promedios de X e Y
X_prom = np.mean(X)
Y_prom = np.mean(Y)
```

```python
# Cálculo del intercepto (A)
A = Y_prom - B * X_prom
```

```python
# Mostrar la ecuación del modelo
print(f"Pendiente (B): {B}")
print(f"Intercepto (A): {A}")
print(f"Modelo: Y = {A:.2f} + {B:.2f}X")
```

```python
# PREDICCIÓN

# Nuevo valor de X para predecir
X_nuevo = 8

# Aplicar el modelo
Y_pred = A + B * X_nuevo

print(f"\nPredicción para X = {X_nuevo}:")
print(f"Y = {Y_pred}")
```
